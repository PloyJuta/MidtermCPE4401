```python
# cmd >> pip install pandas
# cmd >> pip install pandas-datareader
# cmd >> pip install seaborn
# cmd >> pip install matplotlib
import pandas as pd
from pandas_datareader import data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = data.DataReader('fb', data_source="yahoo", start='2010-01-01', end='2021-12-24')
df.head()
```

```python
sns.lineplot(x=df.index, y='Adj Close', data=df);
```

```python
stocks=['fb']
df = data.DataReader(stocks, data_source="yahoo",
                     start="2010-01-01")
df.head()
```

```python
df.index
```

```python
sns.lineplot(x=df.index, y='fb', data=df['Adj Close'])
```

```python
ds=df.stack()
ds
```

```python
ds=df.stack().reset_index()
ds.head()
```

```python
ds[ds.Symbols=='fb']
```

```python
sns.lineplot(x='Date', y='Adj Close', hue='Symbols',
             style='Symbols',
             data=ds);
```

```python
sns.relplot(x='Date', y='Adj Close', col='Symbols', col_wrap=3,
            hue='Symbols',
            kind='line',
            height=5,
            facet_kws={'sharey': False},
            data=ds);
```

```python
sns.lineplot(x='Date', y='Adj Close', hue='Symbols',
             data=ds[ds.Symbols.str.match('fb')]);
```
