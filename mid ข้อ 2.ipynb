```python
import bs4 #BeautifulSoup
from selenium import webdriver #lib ที่ใช้สกัดข้อมูล แต่ความสามารถสูงกว่า ใช้บน window ถ้าเป็น linux จะเป็น dryscrape
#webdriver อยู่ใน selenium
```

```python
browser = webdriver.Chrome('chromedriver.exe') #เรียกใช้โปรแกรมที่เป็น browser จำลอง
url = "https://www.wunderground.com/forecast/VTCR" #แนบลิ้งข้อมูลที่จะดึง
url
```

```python
browser.get(url) #เปิดหน้า web browser #ที่ไม่ใช้ req 
browser.page_source #จะเอาข้อมูลออกมา ต้องใช้ (fn) page_source
```

```python
soup = bs4.BeautifulSoup(browser.page_source,'html.parser') #ทำ string ให้เป็น html
browser.quit() #เปิดแล้วให้ปิดให้ด้วย ถ้าไม่ใส่ต้องปิดเอง
soup
```

```python
soup.prettify() #จัดให้สวยงาม
```

```python
high = soup.find_all('span',{"class":"temp-hi"}) #เก็บข้อมูลอุณหภูมสูงสุดที่อยู่ใน class ='temp-hi'แล้วอยู่ใน tag span มาทั้งหมด ถ้าใช้ find 
#จะดึงมาแค่ตัวแรกที่เจอ และเอาไปใส่ ไว้ใน high
high
```

```python
title_high = []
for h in high:
    title_high.append(h.text)
title_high
```

```python
cool = soup.find_all('span',{"class":"temp-lo"}) #เก็บข้อมูลอุณหภูมสูงสุดที่อยู่ใน class ='temp-lo'แล้วอยู่ใน tag span มาทั้งหมด ถ้าใช้ find 
#จะดึงมาแค่ตัวแรกที่เจอ และเอาไปใส่ ไว้ใน cool
cool
```

```python
title_cool = []
for c in cool:
    title_cool.append(c.text)
title_cool
```
